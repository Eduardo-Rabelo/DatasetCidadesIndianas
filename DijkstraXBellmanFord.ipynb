# Dijkstra para calcular caminho mínimo entre duas cidades

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import time


# Ler o arquivo CSV
nome_arquivo = '/content/indian-cities-dataset.csv' # Substitua pelo nome do seu arquivo CSV
df = pd.read_csv(nome_arquivo)
# Criar um grafo direcionado usando o NetworkX
grafo = nx.DiGraph()

# Adicionar arestas ao grafo a partir do DataFrame

'''
  Nesse for, o _ indica que não mexo no indice das linhas. De resto,
armazeno cada linha na variável row. Crio uma variável pra ser
a cidade de destino, outra pra set a de origem e uma para ser a
distância entre as cidades, que seria o peso das arestas.

  Dessa forma, o grafo é criado pela função add_edge, que cria
vertice com o valor da cidade caso ainda não exista. E  ainda cria
a aresta entre o destino e a origem com o peso sendo a distancia entre
a origem e o destino.
 '''
for _, row in df.iterrows():
    origem = row['Origin']
    destino = row['Destination']
    distancia = row['Distance']
    grafo.add_edge(origem, destino, weight=distancia)


start_vertex = 'Mumbai'
#result = result = nx.single_source_dijkstra(grafo, start_vertex)

start_time =time.time()

result_path= nx.single_source_dijkstra_path(grafo, start_vertex)
result_length = nx.single_source_dijkstra_path_length(grafo, start_vertex)

end_time =time.time()

elapsed_time = start_time-end_time
tempo_djikstra = elapsed_time

print(f"O tempo de execução da parte do algoritmo foi de {tempo_djikstra} segundos.")
print(f"Menor caminho a partir de {start_vertex}: {result_path}")
print(f"Distância do Menor caminho a partir de {start_vertex}: {result_length}")
#print(f"Menor caminho a partir de {start_vertex}: {result}")

nx.draw_shell(grafo, with_labels=True, font_weight="bold")

plt.show()

#Outras formas de plotar o grafo:


# nx.draw_circular(grafo, with_labels=True, font_weight="bold")

# plt.show()

# nx.draw_kamada_kawai(grafo, with_labels=True, font_weight="bold")

# plt.show()

# nx.draw_spectral(grafo, with_labels=True, font_weight="bold")

# plt.show()

# nx.draw_random(grafo, with_labels=True, font_weight="bold")

# plt.show()

# nx.draw_spring(grafo, with_labels=True, font_weight="bold")

# plt.show()

#grafo2=nx.minimum_spanning_tree(grafo)

```

# Teste do Algoritmo de Bellman-Ford para o mesmo Problema

Agora, para o mesmo grafo, o algoritmo de Bellman-Ford será usado para aschar os caminhos mínimos.

```python
start_vertex = 'Mumbai'
#result = result = nx.single_source_bellman_ford(grafo, start_vertex)

start_time =time.time()
result_path= nx.single_source_bellman_ford_path(grafo, start_vertex)
result_length = nx.single_source_bellman_ford_path_length(grafo, start_vertex)
end_time = time.time()

elapsed_time = start_time-end_time
tempo_bellman_ford = elapsed_time
print(f"O tempo de execução da parte do algoritmo foi de {elapsed_time} segundos.")

print(f"Menor caminho a partir de {start_vertex}: {result_path}")
print(f"Distância do Menor caminho a partir de {start_vertex}: {result_length}")
#print(f"Menor caminho a partir de {start_vertex}: {result}")

nx.draw_shell(grafo, with_labels=True, font_weight="bold")

plt.show()

print(f"Tempo Djikstra {tempo_djikstra}")
print(f"Tempo Bellman_Ford: {tempo_bellman_ford}")
```
